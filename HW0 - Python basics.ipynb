## Line Plots

Except for the examples of graphs in the Useful Resources section, below are additional graph types you can plot using matplotlib.

### Scatter Plots

```python
x = np.arange(50) * 2 * np.pi / 50
y = np.sin(x)
fig = plt.figure(figsize=(8, 5)) # create a figure, just like in matlab
ax = fig.add_subplot(1, 1 ,1) # create a subplot of certain size
ax.scatter(x, y, label="y = sin(x)")
ax.set_xlabel('index')
ax.set_ylabel("sin(x)")
ax.set_title("scatter sin(x)")
ax.grid()
ax.legend()
```

### Bar Plots

```python
fig = plt.figure(figsize=(10, 8)) # create a figure, just like in matlab
ax1 = fig.add_subplot(2, 1 ,1) # create a subplot of certain size
ax1.bar(x, y)
ax1.set_xlabel('x')
ax1.set_ylabel("y")
ax1.set_title("bar plot")
ax1.grid()

ax2 = fig.add_subplot(2, 1 ,2) # create a subplot of certain size
ax2.barh(x, y, height=x[1]-x[0])
ax2.set_xlabel('x')
ax2.set_ylabel("y")
ax2.set_title("barh plot")
ax2.grid()

plt.tight_layout()
```

### Exercise 3: Line Plots

Create 1000 values using `linspace` and plot $sin(x) + cos(x)$ both as `plot` and `scatter` in one figure (as 2 different subplots).

```python
# ====== YOUR CODE: ======
x = np.linspace(1, 100, num=1000)
f = np.sin(x)+np.cos(x)

fig = plt.figure(figsize=(10, 8))

ax1 = fig.add_subplot(2, 1 ,1)
ax1.plot(x, f, label="y = sin(x)+cos(x)")
ax1.set_xlabel('x')
ax1.set_ylabel("sin(x)+cos(x)")
ax1.set_title("plot sin(x)+cos(x)")
ax1.grid()
ax1.legend()

ax2 = fig.add_subplot(2, 1 ,2)
ax2.scatter(x, f, label="y = sin(x)+cos(x)")
ax2.set_xlabel('x')
ax2.set_ylabel("sin(x)+cos(x)")
ax2.set_title("scatter sin(x)+cos(x)")
ax2.grid()
ax2.legend()

# ========================
```
